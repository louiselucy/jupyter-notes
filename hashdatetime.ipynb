```python
import pandas as pd
import hashlib
import random
import string
d0 = pd.read_csv(r'C:\Users\louis\OneDrive\merged_goldeikon.csv')
d1 = d0[['datetime','close','amount','ex_symbol']].dropna() 
display(d1.head())
```

```python
tmplist = list(set(d1['datetime']))
```

```python
mapping1 = {i : (''.join(random.choice(string.hexdigits) for i in range(12)))+i+(''.join(random.choice(string.hexdigits) for i in range(12)))  for i in tmplist}
```

```python
d2 = d1.copy()
```

```python
d2['x'] = [mapping1[i] for i in d2['datetime']]
display(d2.head())
```

```python
d2['hash'] = [hashlib.sha1(str.encode(str(i))).hexdigest() for i in d2['x']]
display(d2.head())
```
